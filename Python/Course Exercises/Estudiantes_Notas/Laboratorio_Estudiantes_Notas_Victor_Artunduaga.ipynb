```python
import pandas as pd
import numpy as np

np.random.seed(42)

nombres = ["Ana", "Luis", "Carlos", "María", "Pedro", "Laura", "Andrés", "Camila", "Felipe", "Valentina"]
cursos = ["Matemáticas", "Historia", "Ciencias", "Inglés", "Programación"]
ciudades = ["Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"]
generos = ["Masculino", "Femenino"]

data = {
    "ID": range(1, 201),
    "Nombre": np.random.choice(nombres, 200),
    "Edad": np.random.randint(15, 23, 200),
    "Genero": np.random.choice(generos, 200),
    "Curso": np.random.choice(cursos, 200),
    "Nota1": np.random.randint(50, 101, 200),
    "Nota2": np.random.randint(50, 101, 200),
    "Nota3": np.random.randint(50, 101, 200),
    "Asistencias": np.random.randint(60, 101, 200),
    "Ciudad": np.random.choice(ciudades, 200)
}

df = pd.DataFrame(data)
#crear vriable promedio y obtener el promedio de las notas
df.to_csv("Estudiantes_Notas.csv", index=False)
df.head()
```

```python
df['NotaMax'] = df[['Nota1', 'Nota2', 'Nota3']].max(axis=1)
df
```

```python
df['NotaMin'] = df[['Nota1', 'Nota2', 'Nota3']].min(axis=1)
df
```

```python
promedio_total = df[['Nota1', 'Nota2', 'Nota3']].values.mean()

print("Promedio total:", round(promedio_total,1))
```

```python
df['promedioTotal'] = round(df[['Nota1', 'Nota2', 'Nota3']].mean(axis=1),1)
df
```

```python
promedio_por_curso = df.groupby('Curso')[['Nota1', 'Nota2', 'Nota3']].mean()
promedio_por_curso
promedio_general_curso = promedio_por_curso[['Nota1', 'Nota2', 'Nota3']].mean(axis=1)
round(promedio_general_curso,1)
```

```python
promedio_por_ciudad = df.groupby('Ciudad')[['Nota1', 'Nota2', 'Nota3']].mean()
promedio_por_ciudad
promedio_general_ciudad = promedio_por_ciudad[['Nota1', 'Nota2', 'Nota3']].mean(axis=1)
round(promedio_general_ciudad,1)
```

```python
ciudad_top = promedio_general_ciudad.idxmax()
promedio_top_ciudad = promedio_general_ciudad.max()

print("La ciudad con mayor promedio es:", ciudad_top, "con", round(promedio_top_ciudad,1))

curso_top = promedio_general_curso.idxmax()
promedio_top_curso = promedio_general_curso.max()

print("El curso con mayor promedio es:", curso_top, "con", round(promedio_top_curso,1))
```

```python
!pip install reportlab seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet


doc = SimpleDocTemplate("reporte_estudiantes.pdf", pagesize=letter)
styles = getSampleStyleSheet()
story = []
story.append(Paragraph("REPORTE ANÁLISIS PROMEDIOS DE ESTUDIANTES: VICTOR ARTUNDUAGA", styles["title"]))
story.append(Spacer(1, 12))
intro = """Este reporte presenta un analisis exploratorio de los datos del archivo Notas de estudiantes, se analizan las variables Nota1, Nota2, Nota3, Curso y Ciudad
con el objetivo de comprender patrones entre ellos"""
story.append(Paragraph(intro, styles["Normal"]))
story.append(Spacer(1, 12))
graficos = [("Histograma de Promedios", "grafico1.png", "Este grafico muestra la tendencia de los promedios de todos los estudiantes en la data"),
            ("Promedio de Notas por Curso", "grafico2.png", "Este grafico muestra una comparativa de todos los promedios de los estudiantes que pertenecen a cierto curso en especifico"),
            ("Promedio de Notas por Ciudad", "grafico3.png", "Este grafico de barras horizontales nos presenta de forma sencilla el promedio general de los estudiantes acentados en ciudades especificas"),
            ("Relación de Promedio por Edad","grafico4.png","Se observa como la edad influye en la cantidad de estudiantes presentes en el curso y del mismo modo su promedio general"),
            ("Boxplot del Promedio por Curso","grafico5.png","Este grafico presenta de manera distinta visualmente la relación existente entre el promedio general de los estudiantes que pertenecen a ciertos cursos en especificos y su relación con los demas cursos"),
            ("Mapa de calor de correlaciones","grafico6.png","Este grafico muestra las correlaciones existentes entre las variables pertenecientes al dataframe")]

for titulo, imagen, texto in graficos:
  story.append(Paragraph(f"{titulo}",styles["Heading2"]))
  story.append(Image(imagen, width = 400, height=250 ))
  story.append(Spacer(1,10))
  story.append(Paragraph(texto, styles["Normal"]))
  story.append(Spacer(1,15))
#Resumen
Conclusion = """
Este reporte proporciona un análisis de varias situaciones:
<ul>
<li>Se evidencia una media entre 70 a 75 del promedio total de los estudiantes independiente del curso y/o ciudad</li>
<li>Se evidencian los cursos de programación e Ingles presentan los mejores promedios totales</li>
<li>Se evidencia que los estudiantes que pertenecen a las ciudades de Barranquilla y Cartagena presentan los mejores promedios</li>
</ul>
"""
story.append(Paragraph("Conclusion", styles["Heading2"]))
story.append(Paragraph(Conclusion, styles["Normal"]))
doc.build(story)
print("Reporte generado con exito:reporte_completo.pdf ")
```

```python
import plotly.express as px
import plotly.io as pio
import folium
promedio_ciudad = df.groupby('Ciudad')[['promedioTotal']].mean().reset_index()
Coordenadas = {
    'Bogotá': [4.609710, -74.0],
    'Medellín': [6.244200, -75.578900],
    'Cali': [3.437220, -76.522500],
    'Barranquilla': [10.963890, -74.796389],
    'Cartagena': [10.398980, -75.514440]

}
promedio_ciudad['Latitud'] = promedio_ciudad['Ciudad'].map(lambda x: Coordenadas[x][0])
promedio_ciudad['Longitud'] = promedio_ciudad['Ciudad'].map(lambda x: Coordenadas[x][1])
m = folium.Map(location=[promedio_ciudad['Latitud'].mean(), promedio_ciudad['Longitud'].mean()], zoom_start=5)
for _, row in promedio_ciudad.iterrows():
    folium.CircleMarker(
        location=[row['Latitud'], row['Longitud']],
        radius = 10,
        color = 'blue',
        fill = True,
        fill_color = 'cyan'
    ).add_to(m)

m.save('mapa_promedios_ciudad.html')
m
```
